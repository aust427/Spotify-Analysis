```python
def getAudioFeatures(playlist):
    df = pd.DataFrame(columns=['name', 'danceability', 'energy', 'key', 'loudness',
                           'speechiness', 'acousticness', 'instrumentalness',
                           'liveness', 'valence', 'tempo', 'duration_ms'])
    
    tracks = playlist['items']

    while playlist['next']:
        playlist = sp.next(playlist)
        tracks.extend(playlist['items'])

    for i in range(0, len(tracks)):
        song_features = sp.audio_features(tracks[i]['track']['id'])[0]
        song_features['name'] = tracks[i]['track']['name']

        for key in df.keys():
            df.loc[i,key] = song_features[key]

    df['duration_min'] = df['duration_ms'] / 1000 / 60
    df = df.drop(columns=['duration_ms'])

    return df
```

```python
def plotter(key, bins):
    bin_w = (bins[1] - bins[0]) / 2
    
    mean_l = sum(like_df[key]) / len(like_df[key])
    mean_d = sum(dislike_df[key]) / len(dislike_df[key])
    
    weights_l = np.ones_like(list(like_df[key]))/float(len(like_df[key]))
    weights_d = np.ones_like(list(dislike_df[key]))/float(len(dislike_df[key]))
    
    plt.figure(figsize=(8,8))
    plt.axvline(x=mean_l, color = 'royalblue', ls = '--')
    plt.axvline(x=mean_d, color = 'darkorange', ls = '--')

    hist_l = plt.hist(list(like_df[key]), bins, alpha=0.15, label='like', color = 'royalblue', weights = weights_l)
    hist_d = plt.hist(list(dislike_df[key]), bins, alpha=0.15, label='dislike', color = 'darkorange', weights = weights_d)
    
    xnew = np.arange(bins[0], bins[-2], 0.0001)

    f = interpolate.interp1d(bins[0:len(bins) - 1] + bin_w, hist_l[0], bounds_error=False, kind ='quadratic')
    ynew = f(xnew)
    plt.plot(xnew, ynew, color = 'royalblue')

    f = interpolate.interp1d(bins[0:len(bins) - 1] + bin_w, hist_d[0], bounds_error=False, kind ='quadratic')
    ynew = f(xnew)
    plt.plot(xnew, ynew, color = 'darkorange')

    plt.ylim(ymin = 0)
    plt.legend(loc='upper right')
    plt.title(key)
    plt.show()
```

```python
idx = np.random.choice(like_df.shape[0], dislike_df.shape[0], replace = False)

songs_df = pd.concat([like_df.iloc[idx], dislike_df])
X = songs_df.iloc[:,:-1]
X = X.drop(columns=['name'])
Y = songs_df['like']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25)
```

Evalues k neighbors before assigning a class, usually $k =(n_{row}) ^2$.

## Support Vector Machine (SVM)

SVM draws a hyperplane between points to separate them into classes.

```python
kernels = ["linear", "rbf"]
# "poly",
for i in range(0, 2):
    classifier = SVC(kernel=kernels[i]) 
    classifier.fit(X_train, y_train)
    p = classifier.predict(X_test)
    mat = sklearn.metrics.confusion_matrix(y_true = y_test, y_pred = p)
    class_dict['svm-'+kernels[i]]['matrix'] = mat
    class_dict['svm-'+kernels[i]]['acc'] = (mat[0,0] + mat[1, 1]) / sum(sum(mat))
    class_dict['svm-'+kernels[i]]['classifier'] = classifier
```

```python
def predictor(df, classifier, flag):
    names = df['name']
    df = df.drop(columns=['name'])
    p = classifier.predict(df)
    if (not flag):
        print("accuracy: ", sum(p) / len(p))
    print("labeled dislike:")
    print(names[p == 0])
    if (flag):
        print("labeled like:")
        print(names[p == 1])
```
